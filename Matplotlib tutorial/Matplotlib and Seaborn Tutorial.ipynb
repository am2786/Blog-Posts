```python
x  = np.arange(6)
y1 = x
y2 = x**2
y3 = x**3
```

```python
plt.scatter(x, y1, marker = 's', label = 'x1')
plt.scatter(x, y2, marker = '*', label = 'x2')
plt.scatter(x, y3, marker = '^', label = 'x3')
plt.legend(loc = 'upper left')
```

```python
plt.title('x vs. y', fontsize = 15)
plt.scatter(x, y1, marker = 's', label = 'x1')
plt.scatter(x, y2, marker = '*', label = 'x2')
plt.scatter(x, y3, marker = '^', label = 'x3')
plt.text(1, 100, 'Everyday we lit', fontsize=  20)
plt.legend(loc = 'upper left', fontsize = 15)
plt.xlabel('x', fontsize = 13)
plt.ylabel('y', fontsize = 13);
```

```python
plt.figure(1)

ax1 = plt.subplot(3, 1, 1)
plt.scatter(x, y1, marker = 's', label = 'x1', color = 'b')

ax3 = plt.subplot(3, 1, 2)
plt.scatter(x, y2, marker = '*', label = 'x2', color = 'g')

ax3 = plt.subplot(3, 1, 3)
plt.scatter(x, y3, marker = '^', label = 'x3', color = 'r')
```

```python
plt.figure(figsize = (4,9))
plt.subplots_adjust(hspace=0.0)

ax1 = plt.subplot(3, 1, 1)
plt.scatter(x, y1, marker = 's', label = 'x1', color = 'b')
plt.ylabel('y', fontsize = 13)
plt.axis([0, 6, 0, 140])



ax2 = plt.subplot(3, 1, 2)
plt.scatter(x, y2, marker = '*', label = 'x2', color = 'g')
plt.ylabel('y', fontsize = 13)
plt.axis([0, 6, 0, 140])


ax3 = plt.subplot(3, 1, 3)
plt.scatter(x, y3, marker = '^', label = 'x3', color = 'r')
plt.xlabel('x', fontsize = 13)
plt.ylabel('y', fontsize = 13)
plt.axis([0, 5.5, 0, 140])
;
```
