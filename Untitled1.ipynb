```python
import json
import pandas as pd
from shapely.geometry import Point, Polygon

```

```python
def neighborhood_name_adder(data, geo_data, column_longitude, column_latitude, new_column_name):
    from matplotlib import path
    import numpy as np
    
    
    
    
    lat = data[column_latitude]
    lon = data[column_longitude]
    
    data[new_column_name] = np.zeros(len(data))
    
    for feature in geo_data['features']:
        coords = feature['geometry']['coordinates'][0]
        p = path.Path(coords)
        index = p.contains_points(zip(lon, lat))
        data[new_column_name].loc[index] = [str(feature['properties']['neighborhood'])]*np.sum(index)

    
    return(data)
```

```python
def api_reader(url, geo_path, rows, filename, year):
    
    #This section gets json data
    with open(geo_path) as f:
        geo_data = json.load(f)

    #This section gets taxi api data. Paramater "rows" indicates how many rows of data to get
    url_rows = '?$limit=%s' %rows
    
    url_final = url + url_rows 
    data = pd.read_json(url_final)
    
    #This section adds on the names for the pickup and dropoff neighborhoods
    data_updated = neighborhood_name_adder(data        , geo_data, 'pickup_longitude' , 'pickup_latitude' , 'pickup_neighborhood' )
    data_final   = neighborhood_name_adder(data_updated, geo_data, 'dropoff_longitude', 'dropoff_latitude', 'dropoff_neighborhood')
    
        
    #This section saves it to a csv file
    data_final.to_csv(filename)
```

```python
geo_path = 'NY_neighborhoods.geojson'
rows = 10

url_2011 = 'https://data.cityofnewyork.us/resource/uwyp-dntv.json'
#url_2012 = 'https://data.cityofnewyork.us/resource/kerk-3eby.json'
#url_2013 = 'https://data.cityofnewyork.us/resource/t7ny-aygi.json'
#url_2014 = 'https://data.cityofnewyork.us/resource/gkne-dk5s.json'
#url_2015 = 'https://data.cityofnewyork.us/resource/2yzn-sicd.json'


api_reader(url_2011, geo_path, rows, 'tt.csv', 2011)
```
